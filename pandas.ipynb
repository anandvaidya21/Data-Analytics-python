```python
import numpy as np 
import pandas as pd 
Series1=pd.Series(np.random.random(4))
print(Series1)
print(Series1.index)
```

```python
dict={'m':1,'y':2024,'d':'sunday'}
Series3=pd.Series(dict)
print(Series3)
Series4=pd.Series(dict)
print((Series4))
```

```python
dict={'m':1,'y':2024,'d':'sunday'}
Series3=pd.Series(dict)
print(Series3)
Series4=pd.Series(dict,index=['y','m1','d'])
print((Series4))
```

```python
print('creating series from scaler value')
scl=pd.Series(8,index=['a','b','c','d'])
print(scl)
```

```python
Series1=pd.Series(np.arange(2,10))
print('\n mean:', Series1.mean())
print('\n median:', Series1.median())
print('\n maximum:', Series1.max())
print('\n minimum:', Series1.min())
```

```python
s2=pd.Series((777,88,65,90))
print(s2)
print('\n Name of S2',s2.name)
s2=s2.rename('marks',s2.name)
print(s2)
```
