```python
text="We are using tokeniz@tion in #lp"
symbols=["@","#","|","%"]
current_word= " "
word=[]

for char in text:
    if char.isalnum() or char in symbols:
        current_word+= char
    else:
        if current_word:
            word.append(current_word)
            current_word=''
if current_word:
    word.append(current_word)
print(word)
```
